```python
processed = {
    'block_number': [],
    'collateral': [],
    'debt': [],
    'utilization': [],
    'm_collateral_share': [],
    'm_collateral': [],
    'a_debt_share': [],
    'a_debt': []

    }
for k in total_crv:
    d = total_crv[k]['deposits']
    w = total_crv[k]['withdrawn']
    b = total_crv[k]['borrowed']
    r = total_crv[k]['repayed']
    lc = total_crv[k]['liquidated_collateral']
    ld = total_crv[k]['liquidated_debt']
    total_collateral = d - w - lc
    total_debt = b - r - ld

    md = total_crv[k]['m_deposits']
    mw = total_crv[k]['m_withdrawn']
    mb = total_crv[k]['m_borrowed']
    mr = total_crv[k]['m_repayed']
    mlc = total_crv[k]['m_liquidated_collateral']
    mld = total_crv[k]['m_liquidated_debt']
    m_total_collateral = md - mw - mlc
    m_total_debt =  mb - mr - mld

    ad = total_crv[k]['a_deposits']
    aw = total_crv[k]['a_withdrawn']
    ab = total_crv[k]['a_borrowed']
    ar = total_crv[k]['a_repayed']
    alc = total_crv[k]['a_liquidated_collateral']
    ald = total_crv[k]['a_liquidated_debt']
    a_total_collateral = ad - aw - alc
    a_total_debt =  ab - ar - ald

    processed['block_number'].append(k)
    processed['collateral'].append(total_collateral)
    processed['debt'].append(total_debt)
    processed['utilization'].append(total_debt / total_collateral)
    processed['m_collateral_share'].append(m_total_collateral / total_collateral)
    processed['m_collateral'].append(m_total_collateral)
    processed['a_debt_share'].append(a_total_debt / total_debt)
    processed['a_debt'].append(a_total_debt)
processed
```

```python
df = pd.DataFrame(processed)
df
```

```python
fig = px.bar(df, 
                 x=df.block_number,
                 y=df.collateral,
                title="Total AAVE Collateral: CRV")
fig.update_layout(autotypenumbers='convert types')
fig.show()
```
